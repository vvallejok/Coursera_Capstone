```python
import pandas as pd 
from bs4 import BeautifulSoup
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

```python
print(My_table.tr.text)

headers="Postcode,Borough,Neighbourhood"

table1=""
for tr in My_table.find_all('tr'):
    row1=""
    for tds in tr.find_all('td'):
        row1=row1+","+tds.text
    table1=table1+row1[1:]
print(table1)
```

```python
# Converting data into CSV file
file=open("toronto.csv","wb")

file.write(bytes(table1,encoding="ascii",errors="ignore"))

df = pd.read_csv('toronto.csv',header=None)
df.columns=["Postalcode","Borough","Neighbourhood"]

```

```python
# Converting data into DF
df = pd.read_csv('toronto.csv',header=None)
df.columns=["Postalcode","Borough","Neighbourhood"]
df.head(10)
```

```python
# Get indexes when Borough = NA
indexNA = df[ df['Borough'] =='Not assigned'].index

# Delete indexes from DF
df.drop(indexNA , inplace=True)
df.head(10)
```

```python
#N eighborhood = Borough when Borough = NA
df.loc[df['Neighbourhood'] =='Not assigned' , 'Neighbourhood'] = df['Borough']
df.head(10)
```

```python
# Combine neighborhoods that share postal code into one row and separate them with a comma
SamePC = df.groupby(['Postalcode','Borough'], sort=False).agg( ', '.join)
df_new=SamePC.reset_index()
df_new.head(15)
```

```python
# Numer of rows of the DF
df_new.shape
```

```python
!wget -q -O 'Toronto_long_lat_data.csv'  http://cocl.us/Geospatial_data
df_lon_lat = pd.read_csv('Toronto_long_lat_data.csv')
df_lon_lat.head()
```

```python
df_lon_lat.columns=['Postalcode','Latitude','Longitude']

Toronto_df = pd.merge(df_new,
                 df_lon_lat[['Postalcode','Latitude', 'Longitude']],
                 on='Postalcode')
Toronto_df
```

## Question 3

```python
from geopy.geocoders import Nominatim # address to latitude and longitude values

# importing modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means 
from sklearn.cluster import KMeans

!conda install -c conda-forge folium=0.5.0 --yes 
import folium 

print('Libraries imported.')
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="Toronto")
location = geolocator.geocode(address)
latitude_toronto = location.latitude
longitude_toronto = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude_toronto, longitude_toronto))
```

```python
map_toronto = folium.Map(location=[latitude_toronto, longitude_toronto], zoom_start=10)

# adding markers
for lat, lng, borough, Neighbourhood in zip(Toronto_df['Latitude'], Toronto_df['Longitude'], Toronto_df['Borough'], Toronto_df['Neighbourhood']):
    label = '{}, {}'.format(Neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

```python
num_top_venues = 5

for hood in Toronto_df['Neighborhood']:
    print("----"+hood+"----")
    temp = Toronto_df[Toronto_df['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```
